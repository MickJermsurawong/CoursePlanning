```python
courseFile = open('courseData.txt')
# text = ""
# for each in courseFile:
#     text += each
soup = BeautifulSoup(courseFile,"html.parser")
```

```python
eachItem = soup.findAll("li", { "class" : "item" })

courseNameList = []
courseNameDict = {}
courseCodeList = []
coursePrereq = []
for i, each in enumerate(eachItem):
    playThis = each
    courseNumberTitle = playThis.select("a.trigger")[0].text.strip()
    courseNumberTitle = courseNumberTitle.split()
    courseType = courseNumberTitle[0]
    courseNumber = courseNumberTitle[1]
    courseTitle = " ".join(courseNumberTitle[2:])
    
    courseCodeList.append(courseType+" "+courseNumber)
    courseNameList.append(courseTitle)
    courseNameDict[courseTitle] = i
    prereqs = playThis.select("ul.prerequisites")
    if prereqs:
        prereqs = prereqs[0].select("li")
        thisPrereq = []
        for eachPre in prereqs:
            thisPrereq.append(eachPre.text.strip())
        coursePrereq.append(thisPrereq)
    else:
        coursePrereq.append([])

        
# Abstract class to put
courseNameDict["Introductory Elective in Philosophy"] = len(courseNameList)
courseNameList.append("Introductory Elective in Philosophy")
courseCodeList.append("PHIL-AD 101-119")
coursePrereq.append([])

courseNameDict["History of Philosophy, Theoretical Philosophy, or Practical Philosophy Elective"] = len(courseNameList)
courseNameList.append("History of Philosophy, Theoretical Philosophy, or Practical Philosophy Elective")
courseCodeList.append("PHIL-AD 220-279")
coursePrereq.append([])
```

```python
for i in range(len(courseNameList)):
    print (i, ":", courseCodeList[i], courseNameList[i])
    print (coursePrereq[i])
    for eachP in coursePrereq[i]:
        print (eachP, ":", courseNameDict[eachP])
```
